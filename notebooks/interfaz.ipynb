```python
import streamlit as st
import pandas as pd
import numpy as np

# Título principal
title = "AgroSmart Decisions - Interfaz de Usuario"
st.set_page_config(page_title=title)
st.title(title)

# Menú lateral
st.sidebar.title("Menú Principal")
seccion = st.sidebar.radio("Selecciona una sección:", [
    "Inicio",
    "Cargar Datos",
    "Análisis y Recomendaciones",
    "Visualización",
    "Acerca de"
])

# Sección: Inicio
if seccion == "Inicio":
    st.subheader("Bienvenido a AgroSmart Decisions")
    st.write("Este sistema está diseñado para ayudar a agricultores a tomar decisiones inteligentes basadas en datos.")
    st.image("https://cdn.pixabay.com/photo/2016/07/26/15/28/agriculture-1542518_960_720.jpg", use_column_width=True)

# Sección: Cargar Datos
elif seccion == "Cargar Datos":
    st.subheader("Carga tus datos agrícolas")
    archivo = st.file_uploader("Sube tu archivo CSV", type=["csv"])
    if archivo:
        df = pd.read_csv(archivo)
        st.success("Datos cargados con éxito!")
        st.dataframe(df.head())

# Sección: Análisis y Recomendaciones
elif seccion == "Análisis y Recomendaciones":
    st.subheader("Recomendaciones Inteligentes")
    st.write("Esta sección mostrará recomendaciones personalizadas según los datos cargados.")
    st.info("Funcionalidad en desarrollo. Usa datos simulados por ahora.")
    st.write("Cultivo recomendado: **Tomate** para terrenos con pH entre 6 y 7 y alta materia orgánica.")

# Sección: Visualización
elif seccion == "Visualización":
    st.subheader("Visualización de Datos")
    st.write("Gráficos y tablas para analizar tus parcelas.")
    # Datos simulados
    data = pd.DataFrame({
        "Parcela": ["A", "B", "C", "D"],
        "Materia Orgánica %": [2.5, 3.8, 1.9, 4.2],
        "pH": [6.2, 6.8, 5.5, 7.0]
    })
    st.bar_chart(data.set_index("Parcela"))

# Sección: Acerca de
elif seccion == "Acerca de":
    st.subheader("Sobre este proyecto")
    st.markdown("""
    **Carlos Román**  
    Proyecto MVP10: AgroSmart Decisions  
    Este sistema utiliza análisis de datos para optimizar la producción agrícola de pequeños y medianos agricultores.
    """)
```
